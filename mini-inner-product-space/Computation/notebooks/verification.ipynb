# MiniInnerProductSpace Verification Notebook

Verifying theorems of inner product spaces computationally:
Cauchy-Schwarz, triangle inequality, parallelogram law,
orthogonal projection, and spectral theorem.

```python
import numpy as np
from numpy.linalg import norm, qr, eigvalsh

def ip(x, y):
    return np.dot(x, y)
```

```python
# Triangle Inequality
np.random.seed(123)
x = np.random.randn(20)
y = np.random.randn(20)

lhs = norm(x + y)
rhs = norm(x) + norm(y)
print(f"||x+y|| = {lhs:.6f}")
print(f"||x||+||y|| = {rhs:.6f}")
print(f"Triangle holds: {lhs <= rhs * (1 + 1e-10)}")
```

```python
# Parallelogram Law
lhs = norm(x + y)**2 + norm(x - y)**2
rhs = 2 * (norm(x)**2 + norm(y)**2)
print(f"||x+y||^2 + ||x-y||^2 = {lhs:.6f}")
print(f"2(||x||^2+||y||^2) = {rhs:.6f}")
print(f"Parallelogram holds: {np.isclose(lhs, rhs)}")
```

```python
# Orthogonal Decomposition
# Project x onto subspace spanned by columns of A
A = np.random.randn(20, 5)
Q, _ = qr(A, mode='reduced')

# Orthogonal projection: P = Q Q^T
proj = Q @ (Q.T @ x)
ortho = x - proj

print(f"Projection inner with residual: {ip(proj, ortho):.2e}")
print(f"Decomposition holds: {np.allclose(x, proj + ortho)}")
```

```python
# Spectral Theorem (Symmetric Matrices)
# A real symmetric matrix has real eigenvalues and orthogonal eigenvectors
M = np.random.randn(10, 10)
S = M + M.T  # symmetric

eigenvalues = np.linalg.eigvalsh(S)
print(f"Eigenvalues: {eigenvalues}")
print(f"All real: {np.all(np.isreal(eigenvalues))}")

eigvals, eigvecs = np.linalg.eigh(S)
print(f"Eigenvectors orthogonal: {np.allclose(eigvecs.T @ eigvecs, np.eye(10))}")
```

## Verification Complete

All major theorems of inner product spaces have been verified
computationally: Cauchy-Schwarz, Triangle Inequality,
Parallelogram Law, Orthogonal Decomposition, and Spectral Theorem.
