# MiniInnerProductSpace Exploration Notebook

Exploring inner products, norms, orthogonality, Gram-Schmidt,
and adjoint operators using Python for verification.

```python
import numpy as np
from numpy.linalg import norm, qr

# Inner product / dot product
def inner_product(x, y):
    return np.dot(x, y)

# Norm
def vector_norm(x):
    return np.sqrt(inner_product(x, x))

# Orthogonality check
def is_orthogonal(x, y, tol=1e-10):
    return abs(inner_product(x, y)) < tol
```

```python
# Cauchy-Schwarz verification
np.random.seed(42)
x = np.random.randn(10)
y = np.random.randn(10)

lhs = abs(inner_product(x, y))
rhs = vector_norm(x) * vector_norm(y)
print(f"|<x,y>| = {lhs:.6f}")
print(f"||x||*||y|| = {rhs:.6f}")
print(f"Cauchy-Schwarz holds: {lhs <= rhs}")
```

```python
# Gram-Schmidt orthonormalization
def gram_schmidt(A):
    m, n = A.shape
    Q = np.zeros((m, n))
    for j in range(n):
        v = A[:, j].copy()
        for i in range(j):
            v -= inner_product(Q[:, i], A[:, j]) * Q[:, i]
        Q[:, j] = v / vector_norm(v)
    return Q

A = np.random.randn(5, 3)
Q = gram_schmidt(A)
print("Q^T Q =")
print(Q.T @ Q)
print(f"Orthonormal: {np.allclose(Q.T @ Q, np.eye(3))}")
```

```python
# QR Decomposition
A = np.random.randn(5, 3)
Q, R = qr(A, mode='reduced')
print("Q shape:", Q.shape)
print("R shape:", R.shape)
print(f"A = QR: {np.allclose(A, Q @ R)}")
print(f"Q^T Q = I: {np.allclose(Q.T @ Q, np.eye(3))}")
```
